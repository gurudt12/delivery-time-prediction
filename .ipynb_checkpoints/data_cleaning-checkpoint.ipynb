```python
import pandas as pd
```

```python
df_train = pd.read_excel('dataset/Data_Train.xlsx')
```

```python
df_train.head()
```

```python
df_train.dtypes
```

```python
df_train.describe()
```

```python
def processCost(x):
    try:
        if x[0] == '₹':
            return int(x[1:].replace(',', ''))
        else:
            return int(x).replace(',', '')
    except Exception as e:
        print(e)
        return None
```

```python
df_train.Average_Cost = df_train.Average_Cost.apply(processCost)
```

```python
df_train.Minimum_Order = df_train.Minimum_Order.apply(processCost)
```

```python
df_train.Delivery_Time = df_train.Delivery_Time.apply(lambda x : int(x.split(' ')[0]))
```

```python
df_train.rename(columns = {
    'Delivery_Time' : 'Delivery_Time_min'
}, inplace = True)
```

```python
def processVotes(x):
    if x == '-':
        return 0
    else:
        return int(x)
    
df_train['Votes'] = df_train['Votes'].apply(processVotes)
```

```python
def processReviews(x):
    if x == '-':
        return 0
    else:
        return int(x)
    
df_train['Reviews'] = df_train['Reviews'].apply(processReviews)
```

```python
df_train.head()
```

```python
df_train.dtypes
```

```python
df_train.Cuisines = df_train.Cuisines.apply(lambda x : x.split(', '))
```

```python
df_train.head()
```

```python
unique_cuisines = set()

for cuisines in df_train.Cuisines:
    for cuisine in cuisines:
        unique_cuisines.add(cuisine)
```

```python
len(unique_cuisines)
```

```python
df_train.isna().sum()
```

```python
df_train.dropna(inplace = True)
```

```python
df_train = pd.concat((df_train, pd.DataFrame(columns=unique_cuisines)))
```

```python
df_train.fillna(0, inplace =  True)
```

```python
def processCuisines(df):
    for cuisine in df.Cuisines:
        df[cuisine] = 1
    return df
```

```python
pd.set_option('display.max.columns', None)
```

```python
df_train = df_train.apply(processCuisines, axis=1)
```

```python
df_train.drop('Cuisines', inplace = True, axis=1)
```

```python
df_train.head()
```

```python
df_train.Rating.value_counts()
```

```python
df_train = pd.concat((df_train, pd.DataFrame(columns = ['NEW', 'Opening Soon', 'Temporarily Closed'])))
df_train.fillna(0, inplace = True)
```

```python
def processRatingCases(df):
    rating = df.Rating
    try:
        if rating in ['NEW', 'Opening Soon', 'Temporarily Closed']:
            df[rating] = 1
            return df
        else:
            raise 'int value'
    except:
        return df
```

```python
df_train = df_train.apply(processRatingCases, axis=1)
```

```python
def processRating(x):
    try:
        return float(x)
    except:
        return None
```

```python
df_train.Rating = df_train.Rating.apply(processRating)
```

```python
df_train.fillna('Unknown', inplace = True)
```
